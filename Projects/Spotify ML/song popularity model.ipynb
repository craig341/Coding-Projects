```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import matplotlib

from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
spotify_file_path = '/Users/craig/main/Code/Python/Projects/Spotify ML/dataset.csv'
spotify_data = pd.read_csv(spotify_file_path)

y = spotify_data['popularity']
song_features = ['duration_ms', 'key', 'loudness', 'tempo', 'time_signature', 'speechiness',
                 'acousticness','instrumentalness','liveness','valence','tempo','time_signature']
X = spotify_data[song_features]
```

```python
model = DecisionTreeRegressor(random_state=1)
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)



model.fit(train_X, train_y)
```

```python
predictions = model.predict(val_X)

val_mae = mean_absolute_error(val_y, predictions)
print(val_mae)
```

```python
def get_mae(max_leaf_nodes, train_X=train_X, val_X=val_X, train_y=train_y, val_y=val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)

running = True
max = 1000
while running:
    if get_mae(max)-get_mae(max-1) > 0.01:
        running = False
        print(f"Max leaf nodes: {max}  \t\t Mean Absolute Error: {get_mae(max)}")
    print(f"Max leaf nodes: {max}  \t Mean Absolute Error: {get_mae(max)}")
    max+=1000
```

```python
leaf_nodes = [5, 25, 50, 100, 250, 500, 5000]
mae_list = []

for max_leaf_nodes in leaf_nodes:
     my_mae = get_mae(max_leaf_nodes)
     mae_list.append(my_mae)
     print(f"Max leaf nodes: {max_leaf_nodes}  \t Mean Absolute Error: {my_mae}")
```
